# The TextTiling Algorithm

TheTextTiling algorithm for discovering subtopic structure using termrepetition has three main parts:
1. Tokenization
2. Lexical ScoreDetermination
3. BoundaryIdentification

```python
from enum import Enum
class HYPERPARAMS(Enum):
    w = 20 # token-sequence size
    k = 10 # block size
    n = 1 # number of rounds of smoothing
    s = 2 # smoothing width
```

## Tokenisation

```python
import re
```

```python
class TokenSequence(object):
    "A token list with its original length and its index"

    def __init__(self, index, wrdindex_list, original_length=None):
        original_length = original_length or len(wrdindex_list)
        self.__dict__.update(locals())
        del self.__dict__["self"]


def _divide_to_tokensequences(text, w=20):
    "Divides the text into pseudosentences of fixed size"
    w = HYPERPARAMS.w.value
    wrdindex_list = []
    matches = re.finditer("\w+", text)
    for match in matches:
        wrdindex_list.append((match.group(), match.start()))
    return [
        TokenSequence(i / w, wrdindex_list[i : i + w])
        for i in range(0, len(wrdindex_list), w)
    ]
```

```python
def tokenisation(text):
    lowercase_text = text.lower()
    nopunct_text = "".join(
            c for c in lowercase_text if re.match("[a-z\-' \n\t]", c)
        )
    tokseqs = _divide_to_tokensequences(nopunct_text)
    for ts in tokseqs:
            ts.wrdindex_list = [
                wi for wi in ts.wrdindex_list if wi[0] not in stopwords
            ]
    return tokseqs
```

## Lexical Score Determination

```python
from collections import defaultdict
import math
```

```python
def _words_frequency_map(token_sequences):
    """
    frequency map of words appearing in a block of token sequences
    """
    frequency_map = defaultdict(int)
    for token_sequence in token_sequences:
        for word in [*map(lambda x: x[0], token_sequence.wrdindex_list)]:
            frequency_map[word] += 1
    return frequency_map
```

```python
_words_frequency_map(token_sequences[0:3])
```

```python
def block_comparison_score(i):
    """
    5.2.1 Blocks.
    In the block comparisonalgorithm, adjacentpairs of blocks of token- sequencesare
    comparedfor overall lexical similarity. 
    
    Args:  i (token-sequence gap number, goes from k to len(token_sequences) - k)
    Similarity values are computed for every token-sequence gap number;
    that is, a score is assigned to token-sequence gap i corresponding to how similar the token-sequences 
    from token-sequence i - k to i are to the token-sequence from i + 1 to i + k + 1
    """
    k = HYPERPARAMS.k.value
    first_block = token_sequences[i - k: i + 1]
    second_block = token_sequences[i + 1: i + k + 2]
    
    first_block_frequency_map = _words_frequency_map(first_block)
    second_block_frequency_map = _words_frequency_map(second_block)
    tokens = set(list(first_block_frequency_map.keys()) + list(second_block_frequency_map.keys()))
    
    inner_product = sum([first_block_frequency_map[token] * second_block_frequency_map[token] for token in tokens])
    normalisation = math.sqrt(sum(first_block_frequency_map[token] ** 2 for token in tokens) * sum(second_block_frequency_map[token] ** 2 for token in tokens))
    return inner_product/normalisation
        
```

## Boundary Identification

```python
k = HYPERPARAMS.k.value
block_comparison_score_timeseries = [block_comparison_score(i) for i in range(k, len(token_sequences) - k)]
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(block_comparison_score_timeseries)
plt.title("Block Comparison Score Timeseries (the bigger a valley the more likely a topic change)")
plt.show()
```

```python
def depth_score(timeseries, i):
    """
    The depth score corresponds to how strongly the cues for a subtopic changed on both sides of a 
    given token-sequence gap and is based on the distance from the peaks on both sides of the valleyto that valley.
    
    returns 
    (depth_score, left peak index, right peak index)
    """
    l, r = i - 1, i + 1
    while l > 0 and timeseries[l - 1] > timeseries[l]:
        l -= 1
    while r < (len(timeseries) - 1) and timeseries[r + 1] > timeseries[r]:
        r += 1
    depth_score = (timeseries[r] - timeseries[i]) + (timeseries[l] - timeseries[i])
    return depth_score, l, r
```

```python
depth_score_timeseries = [depth_score(block_comparison_score_timeseries, index)[0] for index, _ in enumerate(block_comparison_score_timeseries[1:-1]) ]
```

```python
plt.plot(depth_score_timeseries)
plt.title("Depth Score Timeseries (how likely is the topic change to be at index X)")
plt.show()
```

```python
biggest_valley_index = depth_score_timeseries.index(max(depth_score_timeseries))
biggest_valley = depth_score(block_comparison_score_timeseries, biggest_valley_index)
```

```python
plt.plot(block_comparison_score_timeseries)
plt.title("Block Comparison Score Timeseries (the bigger a valley the more likely a topic change)")
plt.suptitle("Visualisation of the biggest topic change clue, big red line is the bottom of the valley and the two small red are the peaks")
plt.axvline(biggest_valley[1], c='r')
plt.axvline(biggest_valley[2], c='r')
plt.axvline(biggest_valley_index, c='r', linewidth=5)
plt.show()
```

```python
def smooth(timeseries):
    n, s = HYPERPARAMS.n.value, HYPERPARAMS.s.value
    smoothed_timeseries = timeseries[:]
    for _ in range(n):
        for index, value in enumerate(smoothed_timeseries):
            neighbours = timeseries[max(0, index - s):min(len(timeseries)-1, index +s)]
            smoothed_timeseries[index] = sum(neighbours) / len(neighbours)
    return smoothed_timeseries
    
smoothed_block_comparison_score_timeseries = smooth(block_comparison_score_timeseries)
```
